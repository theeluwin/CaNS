```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
k = 10
B = 100
df = pd.read_pickle('data/game/df_rows.pkl')
stamp_max = df['stamp'].max()
stamp_min = df['stamp'].min()
df['stamp'] = (df['stamp'] - stamp_min) / (stamp_max - stamp_min) * B
df.loc[df['stamp'] > B, 'stamp'] = B
df_tops = df[df['iindex'] <= k][['iindex', 'stamp']]
# bins = np.histogram(df['stamp'], bins=100)[1]
bins = np.histogram(df_tops['stamp'], bins=100)[1]
plt.clf()
for iindex in range(1, k + 1):
    df_cur = df_tops[df_tops['iindex'] == iindex]
    plt.hist(df_cur['stamp'], bins=bins, alpha=0.75, label=f'{iindex}')
plt.legend()
plt.xlabel('Percentage of Absolute Time')
plt.ylabel('Popularity')
plt.show()
```
